```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# creating random data

# create the list of x values using random.rand
x = np.random.rand(5,2)

# Use np.arange to create y values
y = list(np.arange(2,3,0.2))
# x data

print(x, 'length of x is:', len(x))
print(y, 'length of y is:', len(y))
```

```python
# plotting data
# Plotting with plt
plt.plot(x,y, '-r') # general plot command, '-r' (solid red line)
plt.title("Plotting random data") # adding title of the plot
plt.xlabel("x axis values") # adding x_axis label
plt.ylabel("y axis values") # adding y_axis label
plt.grid(True) # adding grid to the plot

# save plot as png
plt.savefig("dacost2_hw1_mlgeo23_fig.png")
```
